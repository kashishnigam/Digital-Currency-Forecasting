```python
import pandas as pd
import yfinance as yf
```

```python
tickerSymbol = 'BTC-USD'
startDate = '2021-01-01'
endDate = '2023-12-31'
btcData = yf.download(tickerSymbol, start=startDate, end=endDate)
```

```python
btcData.head()
```

```python
btcData.tail()
```

```python
exchangeRate = "0.26680"
```

```python
exchangeRate = float(exchangeRate)
```

```python
print("SAR to USD exchange rate:", exchangeRate)
```

```python
columns = btcData.columns

print(columns)
```

```python
columnMapping = {
    'Open': 'Open_USD',
    'High': 'High_USD',
    'Low': 'Low_USD',
    'Close': 'Close_USD',
    'Adj Close': 'Adj Close_USD'
}

btcData.rename(columns=columnMapping, inplace=True)

btcData.head()
```

```python
columns = btcData.columns

print(columns)
```

```python
btcDataUsd = btcData.copy()  
```

```python
btcData.drop(columns='Volume', inplace=True)
```

```python
btcData['Open_SAR'] = btcData['Open_USD'] / exchangeRate
btcData['High_SAR'] = btcData['High_USD'] / exchangeRate
btcData['Low_SAR'] = btcData['Low_USD'] / exchangeRate
btcData['Close_SAR'] = btcData['Close_USD'] / exchangeRate
btcData['Adj Close_SAR'] = btcData['Adj Close_USD'] / exchangeRate
```

```python
btcData.drop(columns=['Open_USD', 'High_USD', 'Low_USD', 'Close_USD', 'Adj Close_USD'], inplace=True)
```

```python
btcData = pd.concat([btcDataUsd, btcData], axis=1)
```

```python
print(btcData.head())
```

```python
columns = btcData.columns

print(columns)
```

```python
btcData.drop(columns=['Adj Close_USD', 'Adj Close_SAR'], inplace=True)
```

```python
column_mapping = {
    'Open_SAR': 'open_SAR', 
    'Open_USD': 'open_USD', 
    'High_SAR': 'high_SAR', 
    'High_USD': 'high_USD', 
    'Low_SAR': 'low_SAR', 
    'Low_USD': 'low_USD', 
    'Close_SAR': 'close_SAR', 
    'Close_USD': 'close_USD', 
    'Volume': 'volume'
}

btcData.rename(columns=column_mapping, inplace=True)

print(btcData.head())
```
